Sumarizando os dados

```python
socios.select(f.year('data_de_entrada_sociedade').alias('ano_de_entrada'))\
      .filter('ano_de_entrada >= 2010')\
      .groupBy('ano_de_entrada').count()\
      .orderBy('ano_de_entrada', ascending=True)\
      .show()
```

```python
#agg = agregate
empresas.select('cnpj_basico', 'porte_da_empresa', 'capital_social_da_empresa')\
        .groupBy('porte_da_empresa')\
        .agg(f.avg('capital_social_da_empresa').alias('capital_social_medio'),
             f.count('cnpj_basico').alias('frequencia')
        )\
        .orderBy('porte_da_empresa', ascending=True)\
        .show()
```

```python
empresas.select('capital_social_da_empresa')\
        .summary()\
        .show()
#summary('count', 'mean', 'stddev', 'min', '25%', '50%', '75%', 'max')
#stddev é media de desvio padrao
```

Pelo schema identificamos que a variável comun é o cnpj_basico, como isso podemos utilizar ela como ligação

```python
empresas_join = estabelecimentos.join(empresas, 'cnpj_basico', how='inner')
```

```python
empresas_join.printSchema()
```

```python
frequencia = empresas_join.select(\
                                  'cnpj_basico', f.year('data_de_inicio_atividade').alias('data_de_inicio'))\
                          .where('data_de_inicio >= 2010')\
                          .groupBy('data_de_inicio')\
                          .agg(f.count('cnpj_basico').alias('freq'))\
                          .orderBy('data_de_inicio', ascending=True)
```

```python
frequencia.toPandas()
```

```python
frequencia.union(
    frequencia.select(
        f.lit('Total').alias('data_de_inicio'),
        f.sum(frequencia.freq).alias('frequencia')
    )
).show()
```

SparkSQl

```python
empresas.createOrReplaceTempView('empresasView')
```

```python
spark.sql("SELECT * FROM empresasView").show(5)
```

```python
spark.sql("""
    SELECT * FROM empresasView
    WHERE capital_social_da_empresa = 50
    """
).show(5)
```

```python
spark.sql("""
      SELECT porte_da_empresa, MEAN(capital_social_da_empresa) AS Media
      FROM empresasView
      GROUP BY porte_da_empresa
""").show(5)
```

```python
empresas_join.createOrReplaceTempView("empresasJoinView")
```

```python
freq = spark.sql("""
      SELECT YEAR(data_de_inicio_atividade) AS data_inicio, COUNT(cnpj_basico) AS count
      FROM empresasJoinView
      WHERE YEAR(data_de_inicio_atividade) >= 2010
      GROUP BY data_inicio
      ORDER BY data_inicio
""")
freq.show()
```

```python
freq.createOrReplaceTempView("freqView")
```

```python
spark.sql("""
      SELECT *
      FROM freqView
      UNION ALL
      SELECT 'Total' AS data_de_inicio, SUM(count) as count
      FROM freqview
""").show()
```
