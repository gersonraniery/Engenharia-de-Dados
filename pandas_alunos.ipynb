# Com base nisso, solucione os problemas propostos abaixo utilizando os conhecimentos adquiridos até aqui.


```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
```

```python
df_alunos = pd.read_csv(url)
```

# 1) Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

```python
df_alunos.isnull()
```

```python
df_alunos.fillna(0)
```

```python
df_alunos = df_alunos.fillna(0)
```

# 2) Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.


```python
alunos_remover = ['Alice', 'Carlos']
```

```python
df_alunos.query('@alunos_remover not in Nome')
```

```python
df_alunos = df_alunos.query('@alunos_remover not in Nome')
```

# 3) Aplique um filtro que selecione apenas os alunos que foram aprovados.

```python
df_alunos.query('Aprovado == True')
```

```python
df_alunos_aprovados = df_alunos.query('Aprovado == True')
```

# 4) Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".


```python
df_alunos_aprovados.to_csv('alunos_aprovados.csv')
```

```python
#testando se salvou corretamente
pd.read_csv('alunos_aprovados.csv')
```

Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade, tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0.

```python
df_alunos['Notas'].replace(7.0, 8.0, inplace=True)
```
